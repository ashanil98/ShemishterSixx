```
def threshhold(img):
  for i in range(len(img)):
    for j in range(len(img[i])):
      if img[i][j] > 200:
        img[i][j] = 255
      else:
        img[i][j] = 0
  return img
```

```
new_img = threshhold(img_arr)
new_img
```

```
new_img = Image.fromarray(np.asarray(new_img, dtype="uint8"), "L")
```
